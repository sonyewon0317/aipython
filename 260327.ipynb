```python
import random

dice1=random.randint(1,6)
dice2=random.randint(1,6)

print(f'주사위1:{dice1}')
print(f'주사위2:{dice2}')

if dice1==dice2:
  print("두 주사위 값이 같다")
elif dice1>dice2:
  print("주사위1값이 주사위2값보다 크다")
else:
  print("주사위2값이 주사위1값보다 크다")
```

```python
#가위(1)바위(2)보(3)
import random

man1=random.randint(1,3)
man2=random.randint(1,3)

print(f'사람1:{man1}')
print(f'사람2:{man2}')

if man1==man2:
  print("두 사람이 비겼다")
elif (man1==1 and man2==2)or(man1==2 and man2==3)or(man1==3 and man2==1):
  print("사람2가 이겼다")
else:
  print("사람1이 이겼다")
```
