```python
# 짝수/홀수 판별 프로그램 (초보자용)

num = int(input("숫자를 입력하세요: "))

# 어떤 수를 2로 나눴을 때 나머지가 0이면 짝수, 1이면 홀수입니다.
if num > 0:
    print("양수입니다.")
else:
    print("음수입니다.")
```
