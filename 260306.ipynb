```python
def calculate(a, b):
    results = {
        "덧셈": a + b,
        "뺄셈": a - b,
        "곱셈": a * b,
        "나눗셈": None
    }

    if b != 0:
        results["나눗셈"] = a / b
    else:
        results["나눗셈"] = "0으로 나눌 수 없습니다."

    return results


if __name__ == "__main__":
    a = float(input("첫 번째 수를 입력하세요: "))
    b = float(input("두 번째 수를 입력하세요: "))

    results = calculate(a, b)

    for op, value in results.items():
        print(f"{op}: {value}")
```
