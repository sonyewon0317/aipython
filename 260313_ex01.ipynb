```python
a = 20  # 첫 번째 숫자를 변수 a에 저장
b = 6   # 두 번째 숫자를 변수 b에 저장

add = a + b  # 덧셈: a와 b를 더한 값을 add에 저장
sub = a - b  # 뺄셈: a에서 b를 뺀 값을 sub에 저장
mul = a * b  # 곱셈: a와 b를 곱한 값을 mul에 저장
div = a / b  # 나눗셈: a를 b로 나눈 값을 div에 저장(결과는 실수일 수 있음)

print("a =", a)          # a 값 출력
print("b =", b)          # b 값 출력
print("a + b =", add)    # 덧셈 결과 출력
print("a - b =", sub)    # 뺄셈 결과 출력
print("a * b =", mul)    # 곱셈 결과 출력
print("a / b =", div)    # 나눗셈 결과 출력
```

```python
a = int(input("첫 번째 숫자 a를 입력하세요: "))  # 사용자에게 a를 입력받고, 정수로 변환해서 저장
b = int(input("두 번째 숫자 b를 입력하세요: "))  # 사용자에게 b를 입력받고, 정수로 변환해서 저장

add = a + b  # 덧셈 결과를 변수 add에 저장
sub = a - b  # 뺄셈 결과를 변수 sub에 저장
mul = a * b  # 곱셈 결과를 변수 mul에 저장
div = a / b  # 나눗셈 결과를 변수 div에 저장(0으로 나누면 오류가 날 수 있음)

print("a =", a)          # 입력받은 a 출력
print("b =", b)          # 입력받은 b 출력
print("a + b =", add)    # 덧셈 결과 출력
print("a - b =", sub)    # 뺄셈 결과 출력
print("a * b =", mul)    # 곱셈 결과 출력
print("a / b =", div)    # 나눗셈 결과 출력
```
