```python
order_counts = {}
total_price = 0

while True:
  print("메뉴판")
  for name, price in cafe_menu.items():
    print(f"{name}: {price}원")
  print()


  menu_name = input("주문(메뉴이름입력) 또는 종료(종료입력)")

if menu_name == "종료":
  break

if menu_name not in cafe_menu:
  print("메뉴판에 없는 메뉴입니다. 다시 입력하세요.")
  continue

qty_str = int(input(f"갯수를 입력하세요.: "))

if not qty_str.isdigit() or int(qty_str) <= 0:
  print("개수는 숫자로 입력해야 합니다. 다시 입력하세요.")
  continue


if menu_name in order_counts:
  order_counts[menu_name] = order_counts[menu_name] + qty
else:
  order_counts[menu_name] = qty

total_price += cafe_menu[menu_name] * qty

print("주문결과")
for name, qty in order_counts.items():
  print(f"{name}: {qty}개")

print(f"총 결제 금액은 {total_price}원입니다.")

```
